```python
import pandas as pd
import numpy as np
```

```python
# Load the data
data = pd.read_csv(r'ber_object_detection.csv')
data
```

```python
data.columns = data.columns.str.strip()
```

```python
# For each SNR value, calculate the average BER for each K value
data_grouped = data.groupby(['SNR', 'K']).mean()
data_grouped = data_grouped.reset_index()
```

```python
# Plot BER vs SNR for each K value
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.semilogy()
for K in data_grouped['K'].unique():
    data_K = data_grouped[data_grouped['K'] == K]
    # Drop K = 128
    if K == 128:
        continue
    plt.plot(data_K['SNR'], data_K['BER'], label=f'K = {K}')
    
plt.grid()
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('BER vs SNR for different K values using object detection')
plt.legend()
plt.show()
```

```python
# Plot BER vs SNR for each K value for background == 1
data_grouped = data.groupby(['SNR', 'K', 'background']).mean()
data_grouped = data_grouped.reset_index()

plt.figure(figsize=(10, 6))
plt.semilogy()
for K in data_grouped['K'].unique():
    data_K = data_grouped[(data_grouped['K'] == K) & (data_grouped['background'] == 1)]
    # Drop K = 128
    if K == 128:
        continue
    plt.plot(data_K['SNR'], data_K['BER'], label=f'K = {K}')
    
plt.xlabel('SNR (dB)')
plt.ylabel('BER')
plt.title('BER vs SNR using object detection: background = 1')
plt.legend()
plt.grid()
plt.show()
    
```
