```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3]
y_bs = [0.44613125920295715, 0.4229322373867035, 0.43116745352745056]
y_ss = [0.860949695110321, 0.7577864527702332, 0.734208345413208]
y_obj = [0.603060245513916, 0.6467483043670654, 0.645075798034668]

label_x = '# of persons'
label_y = 'MSSIM'
```

```python
plt.figure()
plt.plot(x, y_bs, label= 'Baseline')
plt.plot(x, y_ss, label= 'Semantic Segmentation Method')
plt.plot(x, y_obj, label= 'Object Detection Method')
plt.legend(['Baseline', 'Semantic Segmentation Method', 'Object Detection Method'])
plt.xlabel(label_x)
plt.ylabel(label_y)
```
